Анализируем accuracy у **pymorphy**

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
morph.parse(text.split(' ')[int(text.split(' ').index('это'))])
```

```python
def check_pymorphy(text, tagged):
    pymorphy_list = list()
    dictionary_pymorphy = {
        'INFN': 'VERB',
        'PRED': 'VERB',
        'GRND': 'VERB',
        'VERB':'VERB',
        'NOUN': 'NOUN',
        'ADJF': 'ADJ',
        'ADJS': 'ADJ',
        'PRTF': 'ADJ',
        'PRTS': 'ADJ',
        'NUMR': 'NUM',
        'ADVB': 'ADV',
        'COMP': 'ADV',
        'NPRO': 'PRON',
        'PREP': 'ADP',
        'CONJ': 'CONJ',
        'PRCL': 'PART',
        'INTJ': 'PART',  
    }
    for word in range(len(text.split(' '))):
        parsed = morph.parse(text.split(' ')[word])
        for elem in parsed:
            tag = dictionary_pymorphy[str(elem.tag).split(',')[0].split(' ')[0]]
            package = (elem.word, tag)
            if elem == parsed[-1] and package != tagged[word]:
                pymorphy_list.append(package)
            elif package != tagged[word]:
                continue
            else:
                pymorphy_list.append(package)
                break
    counter = 0
    for num in range(len(pymorphy_list)):
        if pymorphy_list[num] != tagged[num]:
            continue
        else:
            counter += 1
    accuracy = counter / 287 * 100
    return accuracy
```

```python
check_pymorphy(text, tagged)
```

Результат выше был получен с учетом некоторой омонимии. Как поведет себя accuracy pymorphy, если мы будем брать только самую первую разметку?

```python
def check_pymorphy_strict(text, tagged):
    pymorphy_list = list()
    dictionary_pymorphy = {
        'INFN': 'VERB',
        'PRED': 'VERB',
        'GRND': 'VERB',
        'VERB':'VERB',
        'NOUN': 'NOUN',
        'ADJF': 'ADJ',
        'ADJS': 'ADJ',
        'PRTF': 'ADJ',
        'PRTS': 'ADJ',
        'NUMR': 'NUM',
        'ADVB': 'ADV',
        'COMP': 'ADV',
        'NPRO': 'PRON',
        'PREP': 'ADP',
        'CONJ': 'CONJ',
        'PRCL': 'PART',
        'INTJ': 'PART',  
    }
    for word in range(len(text.split(' '))):
        parsed = morph.parse(text.split(' ')[word])
        tag = dictionary_pymorphy[str(parsed[0].tag).split(',')[0].split(' ')[0]]
        pymorphy_list.append((parsed[0].word, tag))
    counter = 0
    for num in range(len(pymorphy_list)):
        if pymorphy_list[num] != tagged[num]:
            continue
        else:
            counter += 1
    accuracy = counter / 287 * 100
    return accuracy
```

Довольно неплохо, на самом деле. Теперь анализируем accuracy у **natasha** (точнее, у slovnet)

```python
from navec import Navec
from slovnet import Morph
```

```python
def check_natasha(text, tagged):
    natasha_list = list()
    dictionary_natasha = {
        'VERB': 'VERB',
        'NOUN': 'NOUN',
        'ADP': 'ADP',
        'ADJ': 'ADJ',
        'SCONJ': 'CONJ',
        'CCONJ': 'CONJ',
        'PART': 'PART',
        'ADV': 'ADV',
        'PRON': 'PRON',
        'DET': 'DET',
        'AUX': 'AUX',
        'NUM': 'NUM' 
    }
    morph = Morph.load('slovnet_morph_news_v1.tar', batch_size=4)
    navec = Navec.load('navec_news_v1_1B_250K_300d_100q.tar')
    morph.navec(navec)
    mapped = next(morph.map([text.split(' ')]))
    for token in mapped.tokens:
        tag = dictionary_natasha[str(token.tag).split('|')[0]]
        natasha_list.append((token.text, tag))
    counter = 0
    for num in range(len(natasha_list)):
        if natasha_list[num] != tagged[num]:
            continue
        else:
            counter += 1
    accuracy = counter / 287 * 100
    return accuracy
```

```python
check_natasha(text, tagged)
```

Не так много, как у pymorphy. Наконец, анализируем accuracy у **spacy**

```python
import spacy
```

```python
def check_spacy(text, tagged):
    spacy_list = list()
    dictionary_spacy = {
        'VERB': 'VERB',
        'NOUN': 'NOUN',
        'ADP': 'ADP',
        'ADJ': 'ADJ',
        'SCONJ': 'CONJ',
        'CCONJ': 'CONJ',
        'PART': 'PART',
        'ADV': 'ADV',
        'PRON': 'PRON',
        'DET': 'DET',
        'AUX': 'AUX',
        'NUM': 'NUM' 
    }
    nlp = spacy.load("ru_core_news_sm")
    for word in text.split(' '):
        word_mod = re.sub('-', '', word) #здесь пришлось добавить удаление дефиса, так как спейси как-то плохо работает со словами с дефисом
        for token in nlp(word_mod):
            tag = dictionary_spacy[str(token.pos_)]
            spacy_list.append((token.text, tag))
    counter = 0
    for num in range(len(spacy_list)):
        if spacy_list[num] != tagged[num]:
            continue
        else:
            counter += 1
    accuracy = counter / 287 * 100
    return accuracy
```

```python
check_spacy(text, tagged)
```

```python
def tagged_pymorphy(text):
    dictionary_pymorphy = {
        'INFN': 'VERB',
        'PRED': 'VERB',
        'GRND': 'VERB',
        'VERB':'VERB',
        'NOUN': 'NOUN',
        'ADJF': 'ADJ',
        'ADJS': 'ADJ',
        'PRTF': 'ADJ',
        'PRTS': 'ADJ',
        'NUMR': 'NUM',
        'ADVB': 'ADV',
        'COMP': 'ADV',
        'NPRO': 'PRON',
        'PREP': 'ADP',
        'CONJ': 'CONJ',
        'PRCL': 'PART',
        'INTJ': 'PART',  
    }
    parsed_words = list()
    for word in text.split(' '):
        parsed_word = morph.parse(word)[0]
        tag = dictionary_pymorphy[str(parsed_word.tag).split(',')[0].split(' ')[0]]
        parsed_words.append((parsed_word.word, tag))
    return parsed_words
```

```python
def chunker(tagged_list):
    pos_list = ['VERB', 'NOUN', 'ADP', 'ADJ', 'CONJ', 'PART', 'ADV', 'PRON', 'NUM', 'NEGATION']
    first = str(input('Введите первый параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
    second = str(input('Введите второй параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
    third = str(input('Введите третий параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
    if first not in pos_list or second not in pos_list or third not in pos_list:
        while first not in pos_list or second not in pos_list or third not in pos_list:
            print('Вы ввели параметр, не являющийся доступным. Введите параметры еще раз\n')
            first = str(input('Введите первый параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
            second = str(input('Введите второй параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
            third = str(input('Введите третий параметр (доступные параметры - ' + str(', '.join(pos_list)) + '): ')).upper()
    else:
        answer = list()
        for elem in range(len(tagged_list)):
            if first != 'NEGATION' and second != 'NEGATION' and third != 'NEGATION':
                if tagged_list[elem][1] == first and tagged_list[elem + 1][1] == second and tagged_list[elem + 2][1] == third:
                    answer.append((tagged_list[elem][0], tagged_list[elem + 1][0], tagged_list[elem + 2][0]))
                else:
                    continue
            else:
                if first == 'NEGATION':
                    if tagged_list[elem][0] == 'не' and tagged_list[elem + 1][1] == second and tagged_list[elem + 2][1] == third:
                        answer.append((tagged_list[elem][0], tagged_list[elem + 1][0], tagged_list[elem + 2][0]))
                    else:
                        continue
                elif second == 'NEGATION':
                    if tagged_list[elem][1] == first and tagged_list[elem + 1][0] == 'не' and tagged_list[elem + 2][1] == third:
                        answer.append((tagged_list[elem][0], tagged_list[elem + 1][0], tagged_list[elem + 2][0]))
                    else:
                        continue
                else:
                    if tagged_list[elem][1] == first and tagged_list[elem + 1][1] == second and tagged_list[elem + 2][0] == 'не':
                        answer.append((tagged_list[elem][0], tagged_list[elem + 1][0], tagged_list[elem + 2][0]))
                    else:
                        continue
    return answer
```
